```python
!pip install sweetviz
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sweetviz as sv
df = pd.read_csv('/content/Employee data.csv')
```

```python
# 1. Generate a histogram to visualize the distribution of salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['salary'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Generate a bar plot to compare the average salary of male and female employees
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='salary', data=df, ci=None, palette='viridis')
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.show()
```

```python
# 3. Create a scatter plot to illustrate the relationship between previous work experience and current salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='prevexp', y='salary', data=df, color='green')
plt.title('Relationship between Previous Work Experience and Salary')
plt.xlabel('Previous Work Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# 4. Generate a pie chart to visualize the distribution of educational backgrounds
edu_distribution = df['educ'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(edu_distribution, labels=edu_distribution.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Educational Backgrounds')
plt.show()
```

```python
# 5. Generate a summary of the data using sweetwiz
report = sv.analyze(df)
report.show_html("sweetviz_summary.html")
```
